4. Usa o código do país como uma key no dicionário dict_pais e formata a exibição do value como uma string com a primeira letra em maiúscula e insere o nome de país no dataframe que está sendo analisada.

### Criando as funções que calcula o NAN e o NULL

```python
def quantidade_nan(coluna):
  quantidade_linhas = len(df_final)
  quantidade_de_nan = df_final[coluna].isna().sum()
  porcentagem_nan =  quantidade_de_nan * 100 / quantidade_linhas
  return format(porcentagem_nan, '.2f')
```

```python
def quantidade_null(coluna):
  quantidade_linhas = len(df_final)
  quantidade_de_null = df_final[coluna].isnull().sum()
  porcentagem_null =  quantidade_de_null * 100 / quantidade_linhas
  return format(porcentagem_null, '.2f')
```

```python
def quantidade_vazios(coluna):
  tabela = df_final.loc[df_final[coluna] == '']
  quantidade_vazios = len(tabela)
  quantidade_linhas = len(df_final)
  porcentagem_vazios =  quantidade_vazios * 100 / quantidade_linhas
  return format(porcentagem_vazios, '.2f')
```
